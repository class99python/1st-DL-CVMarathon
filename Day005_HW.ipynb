## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests, json

url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
r = requests.get(url)
data = r.text

jdata = json.loads(data)
print(len(jdata))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for i in range(len(jdata)):
    print(jdata[i]['title'], jdata[i]['createdAt'], 
          jdata[i]['commentCount'], jdata[i]['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'  # 熱門文章
r = requests.get(url)
data = r.text
jdata = json.loads(data)
popAvgCnt = 0
popAvgLike = 0

for i in range(len(jdata)):
    popAvgCnt += jdata[i]['commentCount']
    popAvgLike += jdata[i]['likeCount']
    
print('熱門文章平均留言人數為', popAvgCnt/len(jdata), ', 平均按讚人數為', popAvgLike/len(jdata))


url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=false'  # 最新文章
r = requests.get(url)
data = r.text
jdata = json.loads(data)
popAvgCnt = 0
popAvgLike = 0

for i in range(len(jdata)):
    popAvgCnt += jdata[i]['commentCount']
    popAvgLike += jdata[i]['likeCount']
    
print('最新文章平均留言人數為', popAvgCnt/len(jdata), ', 平均按讚人數為', popAvgLike/len(jdata))
```
