```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests, json

url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
r = requests.get(url)
data = r.text

jdata = json.loads(data)
print(len(jdata))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for i in range(len(jdata)):
    print(jdata[i]['title'], jdata[i]['createdAt'], 
          jdata[i]['commentCount'], jdata[i]['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'  # ç†±é–€æ–‡ç« 
r = requests.get(url)
data = r.text
jdata = json.loads(data)
popAvgCnt = 0
popAvgLike = 0

for i in range(len(jdata)):
    popAvgCnt += jdata[i]['commentCount']
    popAvgLike += jdata[i]['likeCount']
    
print('ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ç‚º', popAvgCnt/len(jdata), ', å¹³å‡æŒ‰è®šäººæ•¸ç‚º', popAvgLike/len(jdata))


url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=false'  # æœ€æ–°æ–‡ç« 
r = requests.get(url)
data = r.text
jdata = json.loads(data)
popAvgCnt = 0
popAvgLike = 0

for i in range(len(jdata)):
    popAvgCnt += jdata[i]['commentCount']
    popAvgLike += jdata[i]['likeCount']
    
print('æœ€æ–°æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ç‚º', popAvgCnt/len(jdata), ', å¹³å‡æŒ‰è®šäººæ•¸ç‚º', popAvgLike/len(jdata))
```
