## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup

url = url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.text[:1000]

soup = BeautifulSoup(r.text, 'html5lib')
print(soup)
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq

g = Grab()
url = url = 'https://www.dcard.tw/f'
r = g.go(url)
r.body[:1000]

doc = pq(r.body)
title = doc('title')
print(title.text())
```
